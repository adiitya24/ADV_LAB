```python
%%R

# Load the dataset (replace 'path_to_your_dataset.csv' with the actual path)
crime_data <- read.csv('crime.csv')

# View the first few rows of the dataset
head(crime_data)
```

```python
%%R
# Bar Chart of property stolen across different areas
barplot(crime_data$Cases_Property_Stolen, names.arg = crime_data$Area_Name, las = 2, col = "skyblue",
        main = "Cases of Property Stolen by Area", xlab = "Area Name", ylab = "Cases of Property Stolen",
        cex.names = 0.7)
```

```python
%%R
# Pie chart for Cases_Property_Stolen in a specific area (e.g., first row)
area_data <- crime_data[1, c('Cases_Property_Stolen')]
labels <- crime_data$Area_Name[4]
pie(as.numeric(area_data), labels = labels, col = rainbow(1), main = "Property Stolen in Specific Area")
```

```python
%%R
# Histogram of Cases_Property_Stolen
hist(crime_data$Cases_Property_Stolen, col = "lightblue", main = "Distribution of Property Stolen Cases",
     xlab = "Number of Property Stolen Cases", breaks = 10)
```

```python
%%R
# Time line chart for Cases_Property_Stolen over the years
plot(crime_data$Year, crime_data$Cases_Property_Stolen, type = "o", col = "red",
     xlab = "Year", ylab = "Number of Cases", main = "Property Stolen Over Time")
```

```python
%%R
# Scatter plot of Cases_Property_Stolen vs Value_of_Property_Stolen
plot(crime_data$Cases_Property_Stolen, crime_data$Value_of_Property_Stolen,
     main = "Cases vs Value of Property Stolen", xlab = "Cases Property Stolen",
     ylab = "Value of Property Stolen", pch = 19, col = "darkgreen")
```

```python
%%R
# Bubble plot of Cases_Property_Stolen vs Value_of_Property_Stolen
symbols(crime_data$Cases_Property_Stolen, crime_data$Value_of_Property_Stolen,
        circles = crime_data$Value_of_Property_Recovered / 1000000,
        inches = 0.5, fg = "red", bg = "lightblue", xlab = "Cases Property Stolen",
        ylab = "Value of Property Stolen", main = "Bubble Plot: Cases vs Value (Size = Value Recovered)")
```
